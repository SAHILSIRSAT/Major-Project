# Creating Time series

```python
import pandas as pd
import numpy as np
from tsfresh import extract_features
```

```python
trs= pd.DataFrame(pd.Series(pd.date_range("2000", freq='0.048828ms', periods=20479)))
```

```python
trs.columns=['time']
```

# Importing and Formatting Healthy State Data

```python
data_healthy=pd.read_csv(r'C:\Users\sahil\Desktop\IMS DATA\Healthy\2004.02.18.15.52.39.txt', delimiter='\t')
data_healthy.columns=['h1', 'h2', 'h3', 'h4']
```

```python
df_raw_healthy = pd.concat([trs, data_healthy], axis=1)
```

```python
df_raw_healthy['id']=pd.DataFrame(np.ones((20479,), dtype=int))
df_raw_healthy_part=df_raw_healthy.iloc[0:200, :]
df_raw_healthy_part.columns
```

```python
finalLIST_healthy=[]
for column in df_raw_healthy_part.columns[1:5]: 
    
    temp_df= df_raw_healthy_part[[column, 'id', 'time']]
    #print(temp_df)
    extracted_features = extract_features(temp_df, column_id="id", column_sort="time")
    #print(extracted_features)
    finalLIST_healthy.append(extracted_features.values.tolist())
    
    #print(finalLIST_healthy)
    temp_df=pd.DataFrame()
    extracted_features=pd.DataFrame()
```

```python
healthy_df = pd.DataFrame(list(zip(finalLIST_healthy[0][0], finalLIST_healthy[1][0], 
                                   finalLIST_healthy[2][0], finalLIST_healthy[3][0])))
final_healthy_Transposed=healthy_df.T
final_healthy_Transposed
```

# Applying Deepinsight

```python
from pyDeepInsight import ImageTransformer, LogScaler
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
X=df_all_states

X
```

```python
%store X 
```

```python
y=pd.DataFrame([0,0,0,0,2,2,2,2,2,2,2,2,3,3,3,3,3,3,3,3])
y.columns=['Labels']
```

```python
%store y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=23, stratify=y)
```

```python
ln = LogScaler()
X_train_norm = ln.fit_transform(X_train)
X_test_norm = ln.transform(X_test)

X_train_norm
```

```python
X_train_norm=X_train_norm.fillna(0)
X_test_norm=X_test_norm.fillna(0)
```

```python
X_test_norm.to_numpy().max()
```

```python

it = ImageTransformer(feature_extractor='pca', 
                      pixels=10, random_state=1701, 
                      n_jobs=-1)
```

```python
plt.figure(figsize=(50, 5))
_ = it.fit(X_train_norm, plot=True)
```

```python
fdm = it.feature_density_matrix()
fdm[fdm == 0] = np.nan

plt.figure(figsize=(10, 7))

ax = sns.heatmap(fdm, cmap="viridis", linewidths=0.01, 
                 linecolor="lightgrey", square=True)
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(5))
for _, spine in ax.spines.items():
    spine.set_visible(True)
_ = plt.title("Features per pixel")
```

```python
px_sizes = [25, (25, 50), 50, 100]

fig, ax = plt.subplots(1, len(px_sizes), figsize=(25, 7))
for ix, px in enumerate(px_sizes):
    it.pixels = px
    fdm = it.feature_density_matrix()
    fdm[fdm == 0] = np.nan
    cax = sns.heatmap(fdm, cmap="viridis", linewidth=0.01, 
                      linecolor="lightgrey", square=True, 
                      ax=ax[ix], cbar=False)
    cax.set_title('Dim {} x {}'.format(*it.pixels))
    for _, spine in cax.spines.items():
        spine.set_visible(True)
    cax.xaxis.set_major_locator(ticker.MultipleLocator(5))
    cax.yaxis.set_major_locator(ticker.MultipleLocator(5))
plt.tight_layout()    
    
it.pixels = 50
```

```python


X_train_img = it.transform(X_train_norm)
```

```python
fig, ax = plt.subplots(1, 15, figsize=(25, 7))
for i in range(0,15):
    ax[i].imshow(X_train_img[i])
    #ax[i].title.set_text("Train[{}] - class '{}'".format(i, y_train[i]))
plt.tight_layout()
```

```python
X_test_img = it.transform(X_test_norm)

fig, ax = plt.subplots(1, 5, figsize=(25, 7))
for i in range(0,5):
    ax[i].imshow(X_test_img[i])
    #ax[i].title.set_text("Test[{}] - class '{}'".format(i, y_test[i]))
plt.tight_layout()
```
